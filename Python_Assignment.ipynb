```python
# Student marks
maths = 85
science = 78
english = 92

# Function to assign grade based on marks
def get_grade(marks):
    if 90 <= marks <= 100:
        return 'A'
    elif 80 <= marks < 90:
        return 'B'
    elif 70 <= marks < 80:
        return 'C'
    elif 60 <= marks < 70:
        return 'D'
    elif 50 <= marks < 60:
        return 'E'
    else:
        return 'F'

# Assigning grades
maths_grade = get_grade(maths)
science_grade = get_grade(science)
english_grade = get_grade(english)

# Displaying results
print("Maths:", maths, "Grade:", maths_grade)
print("Science:", science, "Grade:", science_grade)
print("English:", english, "Grade:", english_grade)
```

Mathematical Programs


```python
num = int(input("Enter a number: "))

if num < 2:
    print("Not a prime number")
else:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            print("Not a prime number")
            break
    else:
        print("Prime number")
```

```python
def find_lcm(x, y):
    greater = max(x, y)
    while True:
        if greater % x == 0 and greater % y == 0:
            return greater
        greater += 1

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("LCM:", find_lcm(a, b))
```

```python
def find_gcd(x, y):
    while y:
        x, y = y, x % y
    return x

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("GCD:", find_gcd(a, b))
```

```python
def calculate(a, b):
    sum_ = a + b
    difference = a - b
    product = a * b
    return sum_, difference, product
result = calculate(10, 5)
print("Sum:", result[0])
print("Difference:", result[1])
print("Product:", result[2])
s, d, p = calculate(10, 5)
print("Sum:", s)
print("Difference:", d)
print("Product:", p)

```

```python
def print_pascals_triangle(n):
    for i in range(n):
        num = 1
        row = []
        for j in range(i + 1):
            row.append(num)
            num = num * (i - j) // (j + 1)
        print(" " * (n - i), " ".join(map(str, row)))

print_pascals_triangle(5)
```
