64.	Identify cities that saw consistent monthly growth for 3+ months in a row.

```python
import pandas as pd

# 1. Prepare the data - ensure proper types
data['OrderDate'] = pd.to_datetime(data['OrderDate'])
data['YearMonth'] = data['OrderDate'].dt.to_period('M')

# 2. Calculate monthly sales per city
monthly_sales = (
    data.groupby(['City', 'YearMonth'])['TotalAmount']
    .sum()
    .unstack(fill_value=0)
    .stack()
    .reset_index(name='MonthlySales')
)

# 3. Sort by City and YearMonth
monthly_sales = monthly_sales.sort_values(['City', 'YearMonth'])

# 4. Calculate month-over-month growth
monthly_sales['Growth'] = (
    monthly_sales.groupby('City')['MonthlySales']
    .pct_change()
    .mul(100)
    .round(2)
)

# 5. Identify growth streaks
def find_growth_streaks(group):
    streaks = []
    current_streak = 0
    for growth in group['Growth']:
        if growth > 0:  # Positive growth
            current_streak += 1
        else:
            if current_streak >= 3:
                streaks.append(current_streak)
            current_streak = 0
    # Check if streak continues to end
    if current_streak >= 3:
        streaks.append(current_streak)
    return pd.Series({
        'MaxStreak': max(streaks) if streaks else 0,
        'GrowthStreaks': streaks,
        'Has3MonthGrowth': any(streak >= 3 for streak in streaks) if streaks else False
    })

# 6. Apply to each city
city_growth = monthly_sales.groupby('City').apply(find_growth_streaks).reset_index()

# 7. Filter cities with 3+ month growth streaks
consistent_growth_cities = city_growth[city_growth['Has3MonthGrowth']]

# 8. Get detailed streak info
def get_streak_details(city):
    city_data = monthly_sales[monthly_sales['City'] == city].set_index('YearMonth').copy() # Set YearMonth as index and use copy
    city_data = city_data.sort_index() # Ensure index is monotonic

    streaks = []
    current_streak = 0
    start_month = None

    # Iterate through the index (YearMonth) and growth values
    for month, growth in city_data['Growth'].items():
        if growth > 0:
            current_streak += 1
            if current_streak == 1:
                start_month = month
        else:
            if current_streak >= 3:
                # Get the previous month Period
                end_month = month.asfreq('M') - 1 # Correct way to get previous period using asfreq and subtraction

                streaks.append({
                    'StartMonth': start_month,
                    'EndMonth': end_month,
                    'Months': current_streak,
                    'AvgGrowth': city_data.loc[start_month:end_month, 'Growth'].mean().round(2)
                })
            current_streak = 0
            start_month = None

    # Check if streak continues to end
    if current_streak >= 3:
        streaks.append({
            'StartMonth': start_month,
            'EndMonth': city_data.index[-1], # Last month in the data
            'Months': current_streak,
            'AvgGrowth': city_data.loc[start_month:, 'Growth'].mean().round(2)
        })

    return streaks

# 9. Create final result
results = []
for _, row in consistent_growth_cities.iterrows():
    city = row['City']
    streaks = get_streak_details(city)
    for streak in streaks:
        results.append({
            'City': city,
            'StartMonth': streak['StartMonth'],
            'EndMonth': streak['EndMonth'],
            'GrowthMonths': streak['Months'],
            'AverageGrowth%': streak['AvgGrowth'],
            'MaxStreak': row['MaxStreak']
        })

growth_report = pd.DataFrame(results).sort_values(['GrowthMonths', 'AverageGrowth%'], ascending=[False, False])

# 10. Display results
print("Cities with 3+ Months of Consistent Growth:")
print(growth_report.to_string(index=False))

# 11. Optional visualization
import matplotlib.pyplot as plt

top_cities = growth_report['City'].unique()[:5]  # Show top 5 for visualization

plt.figure(figsize=(12, 6))
for city in top_cities:
    city_data = monthly_sales[monthly_sales['City'] == city]
    plt.plot(city_data['YearMonth'].astype(str), city_data['MonthlySales'], label=city, marker='o')

plt.title('Monthly Sales Growth in Top Performing Cities')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend()
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```

65.	Generate a YoY (Year-over-Year) growth percentage in TotalAmount for each Region

```python
import pandas as pd

# 1. Prepare the data
data['OrderDate'] = pd.to_datetime(data['OrderDate'])
data['Year'] = data['OrderDate'].dt.year
data['Month'] = data['OrderDate'].dt.month

# 2. Calculate yearly sales by region
yearly_sales = (
    data.groupby(['Region', 'Year'])['TotalAmount']
    .sum()
    .unstack()
)

# 3. Calculate YoY growth
for year in range(yearly_sales.columns.min() + 1, yearly_sales.columns.max() + 1):
    prev_year = year - 1
    if prev_year in yearly_sales.columns:
        yearly_sales[f'YoY_Growth_{year}'] = (
            (yearly_sales[year] - yearly_sales[prev_year]) / yearly_sales[prev_year] * 100
        ).round(2)

# 4. Format results
result = yearly_sales.copy()
result = result.fillna('')  # Remove NaN for display

# 5. Display final YoY growth table
print("Year-over-Year Growth by Region (%):")
print(result.filter(like='YoY').to_string())

# 6. Visualization
import matplotlib.pyplot as plt

# Prepare data for plotting
plot_data = yearly_sales.filter(like='YoY').stack().reset_index()
plot_data.columns = ['Region', 'Year', 'YoY_Growth']
plot_data['Year'] = plot_data['Year'].str.replace('YoY_Growth_', '').astype(int)

plt.figure(figsize=(12, 6))
for region in plot_data['Region'].unique():
    region_data = plot_data[plot_data['Region'] == region]
    plt.plot(region_data['Year'], region_data['YoY_Growth'],
             label=region, marker='o', linestyle='-')

plt.title('Year-over-Year Growth by Region', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Growth Percentage (%)', fontsize=12)
plt.axhline(y=0, color='black', linestyle='--', linewidth=0.5)
plt.grid(axis='y', alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

# 7. Detailed breakdown (optional)
print("\nDetailed Yearly Sales with YoY Growth:")
detailed_view = yearly_sales.copy()
detailed_view = detailed_view.applymap(lambda x: f"${x:,.2f}" if isinstance(x, (int, float)) else x)
print(detailed_view.to_string())
```

89.	After merge, create a master file with these columns: OrderID, OrderDate,
     CustomerName, Product, Category, Price, Quantity, TotalAmount, Returned, and the       AvgProductPrice and TotalCustomerOrders from lookup tables


```python
import pandas as pd

# Load datasets
sales = pd.read_csv('/content/Advanced_Dirty_Sales_Dataset.csv')
customers = pd.read_csv('/content/Advanced_Dirty_Sales_Dataset.csv')
products = pd.read_csv('/content/Advanced_Dirty_Sales_Dataset.csv')

# 1. First, identify common columns for merging
def find_common_col(df1, df2):
    """Find common columns between two DataFrames"""
    common = set(df1.columns) & set(df2.columns)
    return list(common) if common else None

# Find customer join key
cust_join_col = find_common_col(sales, customers)
if not cust_join_col:
    raise KeyError("No common columns found between sales and customers data")
cust_join_col = cust_join_col[0]  # Use first common column

# Find product join key
prod_join_col = find_common_col(sales, products)
if not prod_join_col:
    raise KeyError("No common columns found between sales and products data")
prod_join_col = prod_join_col[0]  # Use first common column

print(f"Merging customers using: {cust_join_col}")
print(f"Merging products using: {prod_join_col}")

# 2. Perform merges
try:
    # Merge sales with customers
    merged = pd.merge(
        sales,
        customers,
        on=cust_join_col,
        how='left'
    )

    # Merge with products
    master = pd.merge(
        merged,
        products,
        on=prod_join_col,
        how='left'
    )

    # 3. Create final columns (with fallbacks)
    final_columns = {
        'OrderID': 'OrderID' if 'OrderID' in master else None,
        'OrderDate': 'OrderDate' if 'OrderDate' in master else None,
        'CustomerName': 'CustomerName' if 'CustomerName' in master else cust_join_col,
        'Product': 'ProductName' if 'ProductName' in master else prod_join_col,
        'Category': 'Category' if 'Category' in master else None,
        'Price': 'Price' if 'Price' in master else 'UnitPrice',
        'Quantity': 'Quantity' if 'Quantity' in master else 'Qty',
        'TotalAmount': 'TotalAmount' if 'TotalAmount' in master else None,
        'Returned': 'Returned' if 'Returned' in master else 'IsReturned',
        'AvgProductPrice': 'AvgProductPrice' if 'AvgProductPrice' in master else None,
        'TotalCustomerOrders': 'TotalCustomerOrders' if 'TotalCustomerOrders' in master else None
    }

    # Filter out None values and keep only existing columns
    available_columns = {k:v for k,v in final_columns.items() if v in master.columns}
    master = master.rename(columns={v:k for k,v in available_columns.items()})[available_columns.keys()]

    # 4. Handle missing data
    for col in ['CustomerName', 'Product', 'Category']:
        if col in master.columns:
            master[col] = master[col].fillna(f'Unknown {col}')

    # Calculate missing numeric fields if needed
    if 'TotalAmount' not in master.columns and all(c in master.columns for c in ['Price', 'Quantity']):
        master['TotalAmount'] = master['Price'] * master['Quantity']

    if 'AvgProductPrice' not in master.columns and 'Price' in master.columns:
        master['AvgProductPrice'] = master.groupby('Product')['Price'].transform('mean')

    if 'TotalCustomerOrders' not in master.columns and 'OrderID' in master.columns:
        master['TotalCustomerOrders'] = master.groupby('CustomerName')['OrderID'].transform('nunique')

    # 5. Save and verify
    print("\nMaster file columns:", master.columns.tolist())
    print("Sample records:")
    print(master.head(3).to_string(index=False))

    master.to_csv('sales_master_file.csv', index=False)
    print("\nMaster file saved successfully!")

except Exception as e:
    print(f"Error during merge: {str(e)}")
    print("\nDebug Info:")
    print("Sales columns:", sales.columns.tolist())
    print("Customer columns:", customers.columns.tolist())
    print("Product columns:", products.columns.tolist())
```
